```python
import pandas as pd

# Load the CSV
df = pd.read_csv("Top selling mobile phones.csv")
df.head()
```

```python
import numpy as np

sales = df['Sales (million units)'].values

print("📊 Average Sales:", np.mean(sales))
print("📈 Maximum Sales:", np.max(sales))
print("📉 Minimum Sales:", np.min(sales))

```

```python
import matplotlib.pyplot as plt

manufacturer_sales = df.groupby('Manufacturer')['Sales (million units)'].sum().sort_values(ascending=False)

manufacturer_sales.plot(kind='bar', figsize=(10, 5), color='orange')
plt.title('Total Sales by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Sales (Million Units)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group sales by manufacturer
manufacturer_sales = df.groupby('Manufacturer')['Sales (million units)'].sum()

# Keep only top 6 manufacturers and group the rest into "Others"
top_sales = manufacturer_sales.sort_values(ascending=False)
top_6 = top_sales[:6]
others = top_sales[6:].sum()

# Add "Others" to the pie
top_6['Others'] = others

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(top_6, labels=top_6.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Market Share by Manufacturer (Sales in %)')
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.show()
```
